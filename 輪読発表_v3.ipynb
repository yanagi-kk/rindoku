# 仮説検定とは
簡単に説明しまーす。<br>
仮説検定とは、「ある仮説を否定(棄却)出来るかどうかを検討したい」だとか、「ある仮説に基づいて、何かしらの意思決定を行いたい(モデル選択等)」といった場合に利用する、判定方法です。

$$
{H_0 : θ ∈ ϴ_0}
$$
$$
{H_1 : θ ∈ ϴ_1}
$$

Hₒのことを「**帰無仮説**」といい、Hₗのことを「 **対立仮説**」という。<br>
また、ϴₒが1点θₒからなる集合であるとき、帰無仮説を「**単純帰無仮説**」といい、「**単純対立仮説**」も同様に定義される。



```
事例１
あるサイコロの１の目が出る確率が確立 p よりも高そう。
そこで、５回そのサイコロを振って、その結果に基づいて「p = 1/6」という仮説を確かめる！！！
```

事例１では、
$$
{H_0 : p = 1/6}
$$
$$
{H_1 : p > 1/6}
$$
として、検定問題を設定できる。ちなみに単純帰無仮説である。

そんでもって、Xの具体的なデータが得られたときに帰無仮説が正しくないとするのが「**棄却**」、正しくないと結論付けられないのが「**受容**」といいます！<br>
でも、検定にはミスがつきものです。ミスのタイプには２種類あり、<br>
本当は帰無仮説が正しいにもかかわらず誤って棄却してしまうことを「**第一種の誤り**」といい、対立仮説が正しいのにもかかわらず誤って受容してしまうことを「**第二種の誤り**」といいます！<br>


この誤りが起きる確率をより小さくするってのが今日のお話。<br>
第一種の誤りをおかす確率と第二種の誤りをおかす確率は相反の関係にあり、双方を同時に小さくすることはできない。<br>


kago.jpg

上の画像で、色がついている面積をできるだけ小さくするのが今回の目標ではあるが、**第一種の誤りの確率αを小さくするために棄却限界値を大きくすると、第二種の誤りの確率βが大きくなってしまう**。逆もしかりである。<br>

そこで伝統的に、**第一種の誤りをおかす確率を一定値α以下に抑えることを第一条件とし、その条件下でなるべく第二種の誤りを小さくする**ように工夫する。

検定δについて、真のパラメータがθの時に帰無仮説を棄却する確率をβ\_δ(θ)で表すと、

$${β_{δ(θ)} ≤ α,　　∀θ∈ϴ_0}$$

となる検定のみを考えて、その中でなるべくいいものを選べばいい！<br>
このαのことを**「有意水準」**といい、一般に1%, 5%, 10%の3つが使われることが多い<br>
第二種の誤りを小さくするということは、対立仮説が正しい時に帰無仮説を棄却する確率をなるべく大きくする、すなわち**ϴₗに属するθについてなるべくβ_δ(θ)を大きくする**ということである。この確率を**「検出力」**と呼び、θの関数であるβ_δ(θ)を**「検出力関数」**という。

Xの値がとりうる空間（標本空間）は、２つに分割される。１つは**「受容域」**と呼ばれる空間で、データがこの空間に値をとった場合、帰無仮説は受容される。もう一つは**「棄却域」**と呼ばれる空間で、こちらにデータの値が属している場合、帰無仮説は棄却される。<br>多くの場合、棄却域はある統計量、すなわちXの関数 T(X) を使って、


$${(x|T(x) > c_1)}, 　{(x|T(x) < c_1)}, $$
$${x|T(x) < c_1 　or　 T(x) > c_2}$$

のような形で表せることが多い。このような場合T(X)を**「検定統計量」**, cᵢ(i=1, 2)を**「棄却限界」**という。cᵢの具体的な値は、有意水準によって決まってくる。

事例１では、X > cの時棄却するので、X自体が検定統計量になり、cが棄却限界になる。<br>よく見られるのは、
$$T(X) = {\frac{(X - μ)}{σ}} $$
$$cを標準正規分布の上側確率の表から求める$$
とするもの。


## p値
検定統計量を使って検定を構成した時に、Xの値が棄却域に属するか、受容域に属するかで判断をくだす代わりに、**p値**と呼ばれるものを計算して棄却か受容かを決めることができる。棄却域が{x | T(x) > c}で与えられている検定δを考えてみる。今、Xの具体的な値がxであったとする。このとき、p値を


$$sup_{θ∈ϴₒ} P(T(X) ≥ T(x))$$

と定義する。p値を計算することで、**αのギリギリで棄却(受容)されたのか、αよりもかなり小さく(大きく)余裕をもって棄却(受容)されたのかがわかる**ので、p値も計算されることはよくある。

## 確率化検定

これまでの話では、検定はXの値だけに基づいて棄却か受容かを判断していたが、場合によっては**Xとは全く別の確率変数Y、特に区間(0, 1)上の一様分布を用いて、XとY双方の値に基づいて判断をする検定方法**を含めて考えた方が、議論しやすくなる。このようなタイプを**「確率化検定」**と呼ぶ。

ここで、次のような確率化検定を考えてみる。

# 一様最強力検定

今回の目的（第一種の誤りの確率が有意水準α以下で最も大きな検定力を持つ検定を見つける）に沿った「最良」の検定として理解しやすいのが、**「一様最強力検定(UMP test)」**であるよって話。<br>
有意水準αの検定δ*がほかのどんな有意水準αの検定δに対しても、

$$
β_δ*(θ) > β_δ(θ),　∀θ∈ϴ_1
$$

を満たすとき、δ*を一様最強力検定と呼ぶ。

これは、対立仮説のどの場合が真の状態であっても、δ\*の検定力を超える検定は存在しないことを意味するので、このような検定を「最良」なものと考えるのは自然なこと。<br>
でも世の中そんな甘くなく、**このような検定は一般には存在しません**。それぞれの検定には検定力の高低に関して得意不得意な θ∈ϴ_1 があり、対立仮説のすべてにおいて、ほかのどの検定も上回るというのは望みえないからでやんす。

じゃあ、**対立仮説が単純仮説**であれば、その特定の対立仮説についてほかのどの有意水準αの検定よりも検定力が高い検定を構成できるんじゃね？って思いますよね。<br>
その答えの一つが、次に述べる**ネイマン-ピアソンの基本定理**ってわけです。

ちなみに、対立仮説が単純仮説の場合は、一様最強力検定の一様をとって、「**最強力検定**」と呼ばれています。

# ネイマン-ピアソンの基本定理



f(x;θ)をパラメータがθのときのXの確率密度関数（離散の時は確率関数）とし、「**１は棄却**」「**０は受容**」「**r は(0, 1)区間の一様分布に従う確率変数 Y が 0 < Y < r の時棄却**」することを意味する。

---
ネイマン-ピアソンの基本定理<br>
**帰無仮説、対立仮説とも単純仮説**で<br><br>
$$
{H_0 : θ = θ_0}　vs.　{H_1 : θ = θ_1}
$$
<br>
の場合、次の形で与えられる検定の**第一種の誤りの確率が α** であれば、この検定は有意水準αの中で**最強力検定**である。
<br><br>
$$
    δ(x) = 
        \begin{cases}
            1 \quad if　f(x;θ_1) / f(x;θ_0) > c \\
            r \quad if　f(x;θ_1) / f(x;θ_0) = c \\
            0 \quad if　f(x;θ_1) / f(x;θ_0) < c \\
        \end{cases}
$$


※他の教科書やWikipediaでは f(x;θ0)/f(x;θ1) < c となってることがあるが同じことを言ってる

---

**f(x;θ0)** : 帰無仮説のもとでの確率関数(X=xの**起こりやすさ**の関数みたいなもん)(**尤度**とも呼ばれる)<br>
**f(x;θ1)** : 対立仮説のもとでの確率関数<br>
**f(x;θ1)/f(x;θ0)** : **尤度比**…**帰無仮説と比べて対立仮説のもとでの確率**が**どれだけ大きいか**の指標みたいなもん<br>

この後に出てくる**尤度比検定との関係性**としては、<br>
- 尤度比検定：
  - 帰無仮説・対立仮説ともに単純仮説である必要はない
    - 下記の L が L > cとなったら棄却するよんってやつ  (P(L > c) < 有意水準α になるから)
$${L = \frac{sup_{θ∈ϴ_1} f(x;θ)}{sup_{θ∈ϴ_0} f(x;θ)}}$$
- ネイマン・ピアソンの基本定理：
  - 帰無仮説・対立仮説ともに**単純仮説**でないといけない
  - そのもとでの**尤度比検定**が有意水準 α の中で最も検定力が出る検定、つまり「**最強力検定**」だよ～ってこと

ネイマン・ピアソン基本定理では単純仮説なので、
$$
{L = \frac{sup_{θ∈ϴ_1} f(x;θ)}{sup_{θ∈ϴ_0} f(x;θ)}} = \frac{f(x;θ_1)}{f(x;θ_0)}$$
となる。

まとめると、ネイマン・ピアソンの基本定理ってのは、<br>
「**尤度比検定の単純仮説バージョン**を利用したら、その検定が**有意水準αの最強力検定**になっちゃうよ、**うれしい**！」ってこと。

ちなみに今までおいらが目指してきた「第一種の誤りの確率 α を決めておき、その中で検出力が最も大きい検定法を選択する」という方針を**「ネイマン・ピアソンの基準」**と呼ぶみたいです。おいらはネイマンとピアソンに従って説明してたみたいですね。。。

---

事例１を例に具体的な使い方を示しまっす。忘れちゃったかもしれないので、**再掲**します。<br>
```
事例１
あるサイコロの１の目が出る確率が確立 p よりも高そう。
そこで、５回そのサイコロを振って、その結果に基づいて「p = 1/6」という仮説を確かめる！！！
```

帰無仮説はp=1/6で、ネイマン・ピアソンは**帰無仮説も対立仮説も両方とも単純仮説**じゃないといけないので、対立仮説はp=1/3とします。

$$
{H_0 : p = \frac{1}{6}}　vs.　{H_1 : p = \frac{1}{3}}
$$

帰無仮説・対立仮説のもとでの確率関数は、

S__12460035.jpg

となり、尤度比はサイコロで１が出た回数 x の関数になる。<br>
したがって、f(x;θ1)/f(x;θ0) > c という形は、x > k という形に書き換えられる。

$$
    δ(x) = 
        \begin{cases}
            1 \quad if　x > k \\
            r \quad if　x = k 　　　(4.2.2)\\
            0 \quad if　x < k \\
        \end{cases}
$$

という形になり、先で見たように **k = 2, r = 0.09024 のとき**、この検定の第一種の誤りの確率 α は 5% になるので、ネイマン・ピアソンの基本定理によりこの(確率化)検定は有意水準α=5%の最強力検定になる。

ここで注意すべきは、今回の検定で選んだ**単純対立仮説 p = 1/3 に依存していない**ことである。**k = 2, r = 0.09024 は有意水準αから決定された数字**で、単純対立仮説 p = 1/3 とは無関係である。<br>
実際に、**1/6より大きいどんな特定の値を単純対立仮説に選んでも**、**結果的に同じ検定が選べる**（k = 2, r = 0.09024になる）。<br>

したがって、この(確率化)検定はパラメータpの真の値が、**対立仮説 H1 : p > 1/6** のどの状況においても、**その検定力が最強**である。つまり、**一様最強力検定**であることがわかる。<br>
このように、**ネイマン・ピアソンの基本定理は、単純帰無仮説 vs 単純対立仮説の場合から、対立仮説が一般的な場合への拡張に応用できる**。<br>
これをさらに、**帰無仮説が一般的な場合にも拡張**したのが次の定理である。この定理の説明のために**単調尤度比**を定義しておく。

### **単調尤度比**
任意の１次元パラメータθ1 < θ2 について、その尤度比 f(x;θ2)/f(x;θ1) が、ある統計量 T(X) を用いて、

$$
  \frac{f(x;θ_2)}{f(x;θ_1)} = g(T(X); θ_1, θ_2)　　　(4.2.2)
$$

にかけたとする。さらに**関数 g(T(X); θ1, θ2)が単調増加**であるとき、Xの確率密度関数(離散型の場合は確率関数)は**T(X)に関して単調尤度比を持つ**という。

---
***単調尤度比と一様最強力検定***<br>
次のような検定問題を考える。
$$
{H_0 : θ ≤ θ_0}　vs.　{H_1 : θ ≥ θ_0}
$$
**Xの確率密度関数（あるいは、確率関数）がT(x)に関して単調尤度比をもつ**とする。このとき、次の形の検定の θ=θ0 のときの棄却の確率がαであれば、この検定は有意水準αの**一様最強力検定**となる。
$$
    δ(x) = 
        \begin{cases}
            1 \quad if　T(X) > c \\
            r \quad if　T(X) = c 　　　(4.2.3)\\
            0 \quad if　T(X) < c \\
        \end{cases}
$$

---

---
つまり、下記の**三つの条件**がそろえば、**一様最強力検定**といえる。
1.  尤度比 f(x;θ2)/f(x;θ1) が、ある統計量 T(x) を用いて(4.2.2)のように書ける。
2.  関数 g(T(X); θ1, θ2)が**単調増加**である。
3.  (4.2.3)の検定の**θ=θ0 のときの棄却の確率がα**である。<br>

---

事例１でいうと、<br>
$$
{H_0 : p ≤ \frac{1}{6}}　vs.　{H_1 : p > \frac{1}{6}}
$$
という問題を考える。

任意の 0 < p1 < p2 < 1 に対して、n=5 の二項分布の確率関数の尤度比は、**①T(x) = xとして、**
$$
\frac{f(x;p_2)}{f(x;p_1)} = \frac{p_2^x(1-p_2)^{5-x}}{p_1^x(1-p_1)^{5-x}} = (\frac{1-p_2}{1-p_1})^5(\frac{p_2(1-p_1)}{p_1(1-p_2)})^{T(x)}
$$
**である。**

右辺はT(x) = x の単調増加関数なので、**②Xは単調尤度比をT(X)に関して持っている**。<br>
したがって、**③(4.2.3)の形の検定のp=1/6の時の棄却の確率がαであれば**、これが**有意水準αの一様最強力検定**になる。

さらに、p2 > p1 より、
$$
\frac{p_2(1-p_1)}{p_1(1-p_2)} > 1
$$
なので、この検定は(4.2.2)の形で書ける。<br>
先にみたように、③**k = 2, r = 0.09024のとき、この検定の p = 1/6 における棄却の確率は5%になる**ので、上記の定理よりこの検定は、**有意水準5%の一様最強力検定**になる。

# 不偏検定

**一般に一様最強力検定は存在しない**が、**検定を自然なクラスに限定する**と、そのクラスの中では一様最強力である検定が存在する場合がある。そのために「**不偏性**」とやらを定義しちゃいましょう。<br>
一般的な検定問題、

$$
{H_0 : θ ∈ ϴ_0}　vs.　{H_1 : θ ∈ ϴ_1}
$$

に対する有意水準αの検定が「**不偏**」であるとは、その検出力関数β(θ)が不等式

$${β_{δ(θ)} ≥ α,　　∀θ∈ϴ_1}$$

を満たすことである。<br>
ここで、**検出力関数β(θ)は帰無仮説を棄却する確率**である。<br>
**棄却すべき状態の時**に、**有意水準よりは高い確率で棄却する**力(検出力)を検定に要求するのは**自然なこと**だよねん。

---
今までは、**帰無仮説のもとで検出力関数β(θ)が有意水準αを超えない**、

$${β_{δ(θ)} ≤ α,　　∀θ∈ϴ_0}$$

どっちかというとこっちを中心に考えて、「**その中で**」検出力が最も大きくなることを考えていた。

---

このようなクラスの中で、一様最強力な検定を**「一様最強力不偏検定 (UMPU test)」**と呼ぶ。すなわち、有意水準αの不偏検定 δ*が、任意の有意水準αの不偏検定 δに対して、


$$
β_δ*(θ) > β_δ(θ),　∀θ∈ϴ_1
$$

であるならば、δ*を一様最強力不偏検定と呼ぶ。

ここで気になる、**「一様最強力不偏検定」になれる条件**ですね。<br>
今回も全体の単調尤度比とやらの時のように、**かなり限定されます**。

それは、**実数確率関数Xが実数パラメータθによって規定される次のような形の確率密度関数(離散型の場合は確率関数)を持つ場合**である。

$$
{f(x;θ) = h(x) exp(θT(x) - c(θ))}　　　(4.2.5)
$$

この形に書ける確率分布を、θを母数(自然母数)とする**指数型分布**と呼ぶ。<br>
(4.2.5)を確率密度関数にもつ分布のθに関して、

$${H_0:θ=ϴ_0　vs.　H_1:θ≠θ_0}$$

という仮説検定問題を、n個の無作為標本 Xi, i=1, 2,...,n に基づいて検定する。これに関して次の定理が成り立つ。

---

**指数分布族と一様最強力不偏検定**<br>

$${\overline{T}(x) = \sum_{i=1}^{n}T(x_i) / n}$$

とし、次のような検定が**①「不偏」**で、**②有意水準がα**であれば、それは**有意水準αの一様最強力不偏検定**になる。

$$
    δ(x) = 
        \begin{cases}
            1 \quad if　\overline{T}(x) > b　or　\overline{T}(x) < a  \\
            r_a \quad if　\overline{T}(x) = a 　　　　　　　　　　(4.2.6)\\
            r_b \quad if　\overline{T}(x) = b \\
            0 \quad if　a < \overline{T}(x) < b \\
        \end{cases}
$$

---

具体例については、後々示す。

要するに、有意水準αで**一様最強力不偏検定になる条件**は
1. Xが**指数型分布族**と呼ばれる分布の種類に属している
2. (4.2.6)の検定が**不偏**である
3. (4.2.6)の検定の**有意水準がα**である

というまたもや限定的なわがまま検定っぷり。<br>
でも、この先に出てくる、「**正規分布に関する検定」「正規分布以外に関する検定」**で**びっくりするほどの活躍**を見せるので期待しててほしい。

**じゃあ、より一般的な条件の下で使える検定方法ってないの？**

# 尤度比検定

ネイマン・ピアソンの基本定理とか、単調尤度比とか、不偏推定のように限定的な条件ではなく、**一般的な条件で使える**のがこの「**尤度比検定**」。多くの具体的な検定方法がこの原理に基づいて導出されている。<br>

次のような一般的な仮説検定問題を考え、

$$
{H_0 : θ ∈ ϴ_0}　vs.　{H_1 : θ ∈ ϴ_1}
$$

Xの確率密度関数(あるいは確率関数)をf(x;θ)としたとき、

$$
L = \frac{sup_{θ∈ϴ_1}f(x;θ)}{sup_{θ∈ϴ_0}f(x;θ)}
$$

を尤度比という。(参考書によっては分母分子を入れ替えた形になることもある)。<br>
尤度比検定は、棄却域が、

$$
L > c
$$

の形で与えられる検定のことである。<br>

検定のゴールとしては、「統計量 T(X) > 具体的な数値」または「T(X) < 具体的な数値」としたときにその確率が有意水準αになることであると考えている。

$$
log L(\hat{θ}) - log L(θ_0)
$$
を変形してLが生まれたと考えてもよい。

この棄却域を具体的な数値として求める方法は２つあり、


① 尤度比 L(θ) > c から得られた棄却の条件が何かの確率分布に従い、その分布から棄却域を具体的な数値として得る<br>
② 帰無仮説のもとで、n(標本の数)が無限に大きくなるとき、
$$
2logL > χ^2_α(p)
$$
を用いて、具体的な数値を得る。

# 漸近分布を使った検定


漸近分布を使った代表的な検定として、**「ワルド検定」**と**「スコア検定」**がある。

両検定ともに、パラメータが他次元な場合にも適用可能な方法であるが、ここではθが1次元の両側検定問題

$$
H_0: θ=θ_0　　vs　　H_1: θ≠θ_0
$$

に話を絞って説明する。

## ワルド検定

$$
n個の無作為標本 X=(X_1, ... ,X_n)が得られたとき、これより作られるθの最尤推定量を\hat{θ}とし、I_F(θ)をθにおける分布におけるフィッシャー情報量とすると、n→∞のとき、帰無仮説のもとで、
$$

$$
W=(\hat{θ}-θ)^2I_F(\hat{θ})\xrightarrow{d}χ^2(1)
$$

となることが知られている。したがって、

$$
W>χ_α^2(1)
$$

とすれば、nが十分に大きいとき、有意水準αの検定が得られる。

例として、

$$
X_i,i = 1,...,nが、正規分布N(0, σ^2)からの無作為標本として、このデータに基づいて検定問題　H_0:σ^2=σ_0^2　vs　H_1:σ^2≠σ_0^2　考えてみる。X = (X_1,...,X_n)の同時確率密度は、
$$

$$
f(x;σ^2)=\frac{1}{(2πσ^2)^\frac{n}{2}}exp(-\frac{1}{2σ^2}\sum_{n=1}^nx_i^2)　　　(4.2.13)
$$

対数をとると、

$$
logf(x;σ^2)=\frac{n}{2}log(2πσ^2)-\frac{1}{2σ^2}\sum_{n=1}^nx_i^2
$$

この式をσ^2で微分すると、

$$
\frac{dlogf(x;σ^2)}{dσ^2}=-\frac{n}{2σ^2}+\frac{1}{2σ^2}\sum_{n=1}^nx_i^2=-\frac{1}{2σ^2}(n-\sum_{n=1}^n(\frac{x_i}{σ})^2)
$$

であり、

$$
I_F(σ^2) = E[(\frac{dlogf(x;σ^2)}{dσ^2})^2] = E[-\frac{1}{4σ^4}(n-\sum_{n=1}^n(\frac{x_i}{σ})^2)^2] = \frac{n}{2σ^4}
$$

となって、(4.2.13)を最大にする最尤推定量は

$$
\hat{σ}^2=\sum_{n=1}^n\frac{x_i^2}{n}
$$

で与えられるので、検定量Wは

$$
W=\frac{n}{2\hat{σ}^4}(\hat{σ}^2-σ_0^2)^2
$$

となる。

便利やね。

## スコア検定

スコア(関数) s(θ) とは対数尤度をパラメータで微分したものである。すなわち、

$$
s(θ) = \frac{dlogf(x;θ)}{dθ}
$$

スコア検定の検定量 R は帰無仮説のもとでの、標本サイズが無限になる条件で、

$$
R=(s(θ_0))^2(I_F(θ_0))^{-1} → χ^2(1)
$$

よって、

$$
R>χ_α^2(1)
$$

とすればよい。

帰無仮説のもとでE(s(θ0))=0となるので、ワルド検定の例の場合のRは、

$$
R=\frac{1}{2n}(n-\sum_{n=1}^n(\frac{x_i}{σ})^2)^2=\frac{n}{2σ_0^4}(\hat{σ}^2-σ_0^2)^2
$$

便利だね。
