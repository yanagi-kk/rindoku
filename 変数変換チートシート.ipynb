# 確率変数が１個の場合
$$
Y = g(X)　⇒　f_{Y}(y)

## 単調増加(減少)の場合
$$
例: Y = σX+μ  　など
$$

まず以下のような変形をする。
$$
X=g^{-1}(Y)
$$

上の式を $ y $ で微分して
$$
\frac{d}{dy}g^{-1}(y)
$$
を求める。

上の二つを以下の公式に突っ込む。
$$
f_{Y}(y) = f_{X}(g^{-1}(y))×\frac{d}{dy}g^{-1}(y)
$$

## まとめ
1. $ X= $ の形にする
2. その式を $ y $ で微分する
3. $ f_Y(y)=f_X(1.の右側)×(2.で求めたもの) $

### 具体例:

#### 問題
$$
f_{X}(x)=λe^{-λx} \\
Y=σX+μ \\
⇩ \\
f_{Y}(y)
$$

#### 解答

まず、$ g^{-1}(Y), \frac{d}{dy}g^{-1}(y) $ を求める。

$$
X=\frac{Y-μ}{σ}=g^{-1}(Y)\\
\frac{d}{dy}g^{-1}(y)=\frac{1}{σ}
$$

より、
$$
f_{Y}(y)=f_{X}(\frac{y-μ}{σ})×\frac{1}{σ}=λ\exp\{-λ(\frac{y-μ}{σ})\}
$$

これだけ。

## 単調増加(減少)じゃない場合
$$
Y = X^2 ,　Y=|X|,　など
$$

分布関数 $ F_{Y}(y) $ を微分して $ f_{Y}(y) $ を求める。

まずは $ F_Y(y) $ を $ F_X(・) $ を使って表現する。
$$
{\begin{align}
F_Y(y)&= P[Y\leq y]\\
&= P[g(X)\leq y]\\
&= P[X\leq h_1(y)]+P[X\leq h_2(y)]\\
&= F_X(h_1(y))+F_X(h_2(y))
\end{align}
}
$$

$ F_X(x) $ を求める。
$$
F_X(x) = \int_{-∞}^{x}f_X(x) dx
$$

$ F_X(x) $ から $ F_Y(y) $ を求め、$ y $ で微分することで $ f_Y(y) $ を求める。

$$
F_Y(y) = F_X(h_1(y))+F_X(h_2(y)) \\
f_Y(y) = \frac{d}{dy}F_Y(y)
$$

### まとめ
- $ F_Y(y) $ を $ F_X(.) $ を使って表す。
- $ F_X(x) $ を求める。
- $ F_Y(y) $ を求めて、$ y $で微分する。

### 具体例:

### 問題
$$
f_{X}(x)=e^{x},　(-∞<x<∞)\\
Y=X^{2} \\
⇩ \\
f_{Y}(y)
$$

### 解答

$$
{\begin{align}
F_Y(y)&= P[Y\leq y]\\
&= P[X^2\leq y]\\
&= P[-\sqrt{y}\leq X \leq\sqrt{y}]\\
&= P[X \leq\sqrt{y}] - P[X \leq -\sqrt{y}]\\
&= F_X(\sqrt{y}) - F_X(-\sqrt{y})
\end{align}
}
$$

$$
{\begin{align}
F_X(x)
&= \int_{-∞}^xe^{t} dt\\
&= [e^{t}]_{-∞}^x\\
&= e^x
\end{align}
}
$$

$$
F_Y(y) = e^{\sqrt{y}}-e^{-\sqrt{y}}
$$

$$
{\begin{align}
f_Y(y)
&= \frac{d}{dy}F_Y(y)\\
&= (e^{\sqrt{y}}-e^{-\sqrt{y}})^{'}\\
&= \frac{1}{2\sqrt{y}}e^{\sqrt{y}}+\frac{1}{2\sqrt{y}}e^{-\sqrt{y}}\\
&= \frac{1}{2\sqrt{y}}(e^{\sqrt{y}}+e^{-\sqrt{y}})
\end{align}
}
$$

以上。

※単調減少と単調増加に分けてから公式を割り当てていると考えるとわかりやすいかも。⇩

$$
{\begin{align}
F_Y(y)&= P[Y\leq y]\\
&= P[X^2\leq y]\\
&= P[-\sqrt{y}\leq X \leq\sqrt{y}]\\
&= P[-\sqrt{y}\leq X \leq 0] + P[0 \leq X \leq \sqrt{y}]\\
&= \{F_X(0)-F_X(-\sqrt{y})\} + \{F_X(\sqrt{y})-F_X(0)\}\\
&= F_X(\sqrt{y}) - F_X(-\sqrt{y})
\end{align}
}
$$

# 確率変数が２個の場合
$$
\left\{
\begin{array}{ll}
Z=z(X, Y) \\
W=w(X, Y)
\end{array}
\right.
$$

まず、以下のような変形を行う。

$$
{\begin{cases}
Z=z(X, Y) \\
W=w(X, Y)
\end{cases}\Leftrightarrow
\begin{cases}
X=x(Z, W) \\
Y=y(Z, W)
\end{cases}
}
$$

## $ (Z, W) ⇒(X, Y)$  が一対一関係の場合

$$
例: \\
X, Yは互いに独立で、X, Y～exp(λ)\\
{\begin{cases}
Z=X+Y \\
W=Y
\end{cases}\Leftrightarrow
\begin{cases}
X=Z-W \\
Y=W
\end{cases}
}
$$

$ ヤコビヤン J((z, w) ⇒ (x, y))$ を求める。
$$
J((z, w) ⇒ (x, y))=
det\begin{pmatrix}
\frac{∂x}{∂z} & \frac{∂x}{∂w} \\
\frac{∂y}{∂z} & \frac{∂y}{∂w}
\end{pmatrix}
$$

#### ヤコビヤンのコツ
- $ (z,w) $ を見て「左の列が $ ∂z、 $ 右の列が $ ∂w $ 」と覚えると良い。あとは普通に考えてxが上、yが下。
- $x,y$ で微分するか、$z,w$ で微分するかわからなくなる人は、もう上の書き方をコピーして、$z,w$ で微分してください。
- 上でやると、ヤコビヤンを使うときに逆数をとる必要がない(ただかけるだけ)なのでわかりやすいです。

$ f_{Z,W}(z,w) $をヤコビヤンを使って求める。

$$
{\begin{align}
f_{Z,W}(z,w)
&= f_{X,Y}(x(z,w),y(z,w))×|J|\\
&= f_X(x(z,w))×f_Y(y(z,w))×|J|　(X,Y は互いに独立)
\end{align}
}
$$

※ $ X,Y $ は大体互いに独立なので上のように書いた。一応問題文を見て互いに独立か確認する。(互いに独立じゃない問題を見たことない。)

$ w $ の範囲を求める。
$$
上の例だと、\\
(0 ≤ z-w ≤ ∞,　0 ≤ w ≤ ∞)\\
\Leftrightarrow
(w ≤ z ≤ ∞,　0 ≤ w ≤ ∞)
$$

より、
$$
0≤w≤z
$$
となる。

$ w $ の範囲で $ f_{Z,W}(z,w) $ を積分して $ f_Z(z) $ を求める。

$$
f_Z(z) = ∫f_{Z,W}(z,w) dw
$$

### まとめ
- $ X=, Y= $の形に変形する
- $ ヤコビヤン J((z, w) ⇒ (x, y))$ を求める。
- $ f_{Z,W}(z,w) $をヤコビヤンを使って求める。
- $ w $ の範囲を求める。　⇐ 忘れがち！！！
- $ w $ の範囲で $ f_{Z,W}(z,w) $ を積分して $ f_Z(z) $ を求める。

### 具体例

### 問題

$$
X, Yは互いに独立で、X, Y～N(0,1)\\
{\begin{cases}
Z=X+Y \\
W=Y
\end{cases}
}
$$

### 解答

$$
{\begin{cases}
Z=X+Y \\
W=Y
\end{cases}\Leftrightarrow
\begin{cases}
X=Z-W \\
Y=W
\end{cases}
}
$$

$$
{\begin{align}
J((z, w) ⇒ (x, y))
&= det\begin{pmatrix}
\frac{∂x}{∂z} & \frac{∂x}{∂w} \\
\frac{∂y}{∂z} & \frac{∂y}{∂w}
\end{pmatrix}\\
&= det\begin{pmatrix}
1 & -1 \\
0 & 1
\end{pmatrix}\\
&= 1
\end{align}
}
$$

$$
{\begin{align}
f_{Z,W}(z,w)
&= f_X(z-w)×f_Y(w)×1\\
&= \frac{1}{\sqrt{2π}}exp\{-\frac{(z-w)^2}{2}\} × \frac{1}{\sqrt{2π}}exp(-\frac{w^2}{2})\\
&= \frac{1}{2π}exp\{-\frac{(z-w)^2+w^2}{2}\}
\end{align}
}
$$

$$
(-∞≤z-w≤∞,　-∞≤w≤∞)\\
\Leftrightarrow(-∞≤z≤∞,　-∞≤w≤∞)\\
$$

$$
-∞≤w≤∞
$$

$$
{\begin{align}
f_Z(z)
&= \int_{-∞}^∞\frac{1}{2π}exp\{-\frac{(z-w)^2+w^2}{2}\}dw ⇒方針: wについての正規分布の確率密度関数の形にして1で消す\\
&= \frac{1}{2π}\int_{-∞}^∞exp\{-\frac{2w^2-2zw+z^2}{2}\}dw\\
&= \frac{1}{2π}\int_{-∞}^∞exp\{-\frac{2(w-\frac{z}{2})^2-\frac{z^2}{2}+z^2}{2}\}dw\\
&= \frac{1}{\sqrt{2}}\frac{1}{\sqrt{2π}}exp(-\frac{z^2}{2})\int_{-∞}^∞\frac{1}{\sqrt{2π(\frac{1}{\sqrt{2}})^2}}exp\{-\frac{(w-\frac{z}{2})^2}{2×(\frac{1}{\sqrt{2}})^2}\}dw\\
&= \frac{1}{\sqrt{2}}\frac{1}{\sqrt{2π}}exp(-\frac{z^2}{2})
\end{align}
}
$$
